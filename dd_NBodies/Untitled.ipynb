```python
import matplotlib.pyplot as plt
import numpy as np


import json
with open('log.json') as f:
  data_energy = json.load(f)

figure, axes = plt.subplots(figsize = (20,20))

axes.plot(list(map(float, data_energy["n"])),list(map(float, data_energy["energy"])),linestyle='solid', marker='o', markersize=4, label='dE/E0')

axes.set_xlabel('n computes')
axes.set_ylabel('|E-E0|/E0')
axes.legend()

font = { 'size'   : 40}

plt.rc('font', **font)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
import numpy as np

import json
with open('bodies.json') as f:
  data_bodies = json.load(f)

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')

for item in data_bodies:
    ax.plot(list(map(float, item["X"])),list(map(float, item["Y"])),list(map(float, item["Z"])))

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

#ax.legend()
plt.savefig('anime2.png', dpi=300)
```
